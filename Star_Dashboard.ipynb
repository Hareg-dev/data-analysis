```python
# calculate summary statistics
mean = data['Luminosity(L/Lo)'].mean()
std  = data['Luminosity(L/Lo)'].std()

# Calculate the lower and upper bounds
lower_bound = mean - std*2
upper_bound = mean + std*2

print('Lower Bound :',lower_bound)
print('Upper Bound :',upper_bound)

# Drop the outliers
data2= data[(data['Luminosity(L/Lo)'] >= lower_bound)
                & (data['Luminosity(L/Lo)'] <= upper_bound)]
data2.describe()
```

```python
# calculate summary statistics
mean = data['Temperature (K)'].mean()
std  = data['Temperature (K)'].std()

# Calculate the lower and upper bounds
lower_bound = mean - std*2
upper_bound = mean + std*2

print('Lower Bound :',lower_bound)
print('Upper Bound :',upper_bound)

# Drop the outliers
data2= data[(data['Temperature (K)'] >= lower_bound)
                & (data['Temperature (K)'] <= upper_bound)]
data2.describe()
```

```python
# @title Temperature (K)

from matplotlib import pyplot as plt
data2['Temperature (K)'].plot(kind='line', figsize=(8, 4), title='Temperature (K)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
data['Log_Luminosity'] = data['Luminosity(L/Lo)'].apply(lambda x: np.log10(x) if x > 0 else 0)  # Log scale for luminosity


sns.set(style='darkgrid')


fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Star Classification Dashboard', fontsize=16)

sns.scatterplot(data=data, x='Temperature (K)', y='Absolute magnitude(Mv)',hue='Star type',
                palette='deep', ax=axes[0, 0])
axes[0, 0].invert_xaxis()
axes[0, 0].set_title('Hertzsprung-Russell Diagram')

sns.histplot(data=data, x='Temperature (K)',hue='Star type', multiple='stack', ax=axes[0, 1])
axes[0, 1].set_title('Temperature Distribution by Star Type')

sns.boxplot(data=data, x='Star type', y='Log_Luminosity', ax=axes[1, 0])
axes[1, 0].set_title('Log Luminosity by Star Type')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(data=data, x='Star type', ax=axes[1, 1])
axes[1, 1].set_title('Star Type Counts')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
data['Temperature (K)'] = np.log(data['Temperature (K)'] + 1)

# Plot original vs transformed
sns.histplot(data['Temperature (K)'], kde=True, color='blue', label='Original')
#sns.histplot(data['log_Temperature (K)'], kde=True, color='green', label='Log Transformed')
plt.legend()
plt.show()
```
